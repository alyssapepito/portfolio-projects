# Data Science Job Postings on Glassdoor

In this project, I had data about <a href="https://www.kaggle.com/datasets/rashikrahmanpritom/data-science-job-posting-on-glassdoor/data" target="_blank">Data Science Job Postings on Glassdoor</a>.

I wanted to practice data cleaning and transformation techniques using Python. I was able to remove duplicates, drop unnecessary columns, create new columns from the original data, fill null values, and standardizing column values using replace.

```python
# Import packages
import pandas as pd
import numpy as np

# Read data
df = pd.read_csv(r"C:\Users\User\Portfolio Projects\Data Science Job Postings Project\Uncleaned_DS_jobs.csv")
df
```

```python
# Check for missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Check for duplicates
duplicates_check = df.duplicated().any()
duplicates_check

# Remove if there are duplicates
# df = df.drop_duplicates()
# df
```

```python
# Check the data types of the columns
print(df.dtypes)
```

```python
# Clean the 'Salary Estimate' column
char_to_replace = {'K': '', '(Glassdoor est.)': '', '(Employer est.)': '', '$':''}

# Iterate over all key-value pairs in dictionary 
for key, value in char_to_replace.items():
    # Replace key character with value character in string
    df['Salary Estimate'] = df['Salary Estimate'].str.replace(key, value)

df
```

```python
# Create two columns from the 'Salary Estimate' column to differentiate minimum and maximum salaries
df[['min_salary', 'max_salary']] = df['Salary Estimate'].str.split('-', n=2, expand=True)

# Make a new column as the average salary using the mean of 'min_salary' and 'max_salary' columns
df[['min_salary', 'max_salary']] = df[['min_salary', 'max_salary']].astype(int)
df['avg_salary'] = df[['min_salary', 'max_salary']].mean(axis=1).astype(int)

df
```

```python
# Clean the 'Company Name' column
df['Company Name'] = df['Company Name'].str.replace(r'\n[0-9]\.[0-9]', '', regex=True)
df
```

```python
# Create a state column from the 'Location' column
states = {'California':'CA', 'New Jersey':'NJ', 'Texas':'TX', 'Utah':'UT'}

df.loc[:, 'job_state'] = df['Location'].str.split(',', expand=True)[1]
df['job_state'] = df['job_state'].fillna(df['Location'].map(states))
df
```

```python
# Create a column that states if location and headquarters are in the same place
df.loc[:, 'same_loc'] = np.where(df['Location'] == df['Headquarters'], True, False)
df
```

```python
# Create a column that states the company age using the 'Founded' column
df.loc[:, 'company_age'] = np.where(df['Founded'] != -1, 2024 - df['Founded'], -1)
df
```

```python
# Create columns that indicate if python, excel, hadoop, spark, aws, tableau, and big data were mentioned in the 'Job Description' column
df.loc[:, 'python'] = np.where(df['Job Description'].str.contains(r'(?i)\bpython\b'), True, False)
df.loc[:, 'excel'] = np.where(df['Job Description'].str.contains(r'(?i)\bexcel\b'), True, False)
df.loc[:, 'hadoop'] = np.where(df['Job Description'].str.contains(r'(?i)\bhadoop\b'), True, False)
df.loc[:, 'spark'] = np.where(df['Job Description'].str.contains(r'(?i)\bspark\b'), True, False)
df.loc[:, 'aws'] = np.where(df['Job Description'].str.contains(r'(?i)\baws\b'), True, False)
df.loc[:, 'tableau'] = np.where(df['Job Description'].str.contains(r'(?i)\btableau\b'), True, False)
df.loc[:, 'big_data'] = np.where(df['Job Description'].str.contains(r'(?i)\bbig data\b'), True, False)
df
```

```python
# Create a column that indicates if the job requires seniority
df.loc[:, 'seniority'] = np.where(df['Job Title'].str.contains(r'(?i)\bsenior|sr\b'), True, False)
df
```

```python
# Drop unnecessary columns
df = df.drop(columns = ['index', 'Founded', 'Competitors'])

# Reset index
df = df.reset_index(drop=True)
df
```

```python
# Saving the dataframe into a CSV
df.to_csv(r"C:\Users\User\Portfolio Projects\Data Science Job Postings Project\Cleaned_DS_jobs.csv", index=False)
```
